```python
import matplotlib.pyplot as plt

#L=300 TANH H1:8, H2:5 n=0.001
with open('mlp_result_points.txt', 'r') as file:
    for line in file:
        tmp = line.split(",")
        if(tmp[2].rstrip()  == '+'):
            plt.scatter(float(tmp[0]), float(tmp[1]), marker ='+', color = 'r')
        elif(tmp[2].rstrip()  == '-'):
            plt.scatter(float(tmp[0]), float(tmp[1]), marker ='_', color = 'b')

        
```
